```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 200)
```

```python
df_application_train = pd.read_csv("../data/application_train.csv")
```

```python
df_application_train.shape
df_application_train.info()

df_application_train.describe().T
```

```python
target_counts = df_application_train['TARGET'].value_counts()

target_percent = df_application_train['TARGET'].value_counts(normalize=True) * 100

summary = pd.DataFrame({
    'Count': target_counts,
    'Percent': target_percent.round(2)
})

print(summary)
```
